```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]

# minX = min(cxMarks)
# minY = min(cyMarks)
# minValue = min(minX,minY)

# write a code to generate below graph
plt.plot(x,cxMarks,marker='>',linestyle='dashed',c='r',label='CX')
plt.xlabel("Roll No")
plt.plot(x,cyMarks,marker='o',linestyle='dotted',c='b',label='CY')
plt.ylabel("CPI")

plt.annotate("Lowest CPI",(8.3,1))
plt.legend()
plt.show()
```

### 04) WAP to demonstrate the use of Pie chart.

```python
dept = ["CSE","ME","EE","EC","AE","CE"]
students = [120,50,80,100,20,30]
plt.pie(students,labels=dept)
plt.legend(title="Engineering Department",loc="upper right")
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
dept = ["CSE","ME","EE","EC","AE","CE"]
students = [120,50,80,100,20,30]
plt.bar(dept,students)
plt.xlabel("Branch")
plt.ylabel("Students")
plt.title("Contribution by Branch")
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
dept = ["CSE","ME","EE","EC","AE","CE"]
students = [120,50,80,100,20,30]
bars = plt.bar(dept,students)

for i in bars:
    yc = i.get_height()
    plt.text(i.get_x()+i.get_width()/2,yc+1,yc,ha='center')

plt.xlabel("Branch")
plt.ylabel("Students")
plt.title("Contribution by Branch")
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
colors = ["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"]
plt.scatter(x,y ,color = colors)
plt.show()
```
