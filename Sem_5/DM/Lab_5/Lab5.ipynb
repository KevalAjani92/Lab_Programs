### 2)  Handle Missing Values in data set [use dropna(), fillna(), and interpolate]


```python
data_withdropna = df.dropna(how='any',axis=1)   # will drop columns which have null value
data_withdropna
# how = 'all' => if all values of row are null then it will be drop
# how = 'any' => Will Drop All Row wich has any one null value
# how = 'any',axis = 1 => drop columns (default aixs = 0 measn row)
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
# Min-Max
data_scaled = df.copy()
# data_scaled['Age'] = data_scaled['Age'].interpolate()
min_age = data_scaled['Age'].min()
max_age = data_scaled['Age'].max()
data_scaled['Age_scaled'] = (data_scaled['Age'] - min_age)/(max_age-min_age)
data_scaled
```

```python
# Decimal Scaling
j = len(str(int(max_age)))
data_scaled['Age_scaled'] = df['Age']/10**j
data_scaled
```

```python
# Z-score
mean_age = df['Age'].mean()
std_age = df['Age'].std()
data_scaled['Age_scaled'] = (data_scaled['Age']-mean_age)/std_age
data_scaled
```
