### Step 19. What is the average revenue amount per order?

```python
# Solution 1
chipo['item_revenue'] = chipo['quantity']*chipo['item_price']
orderwise_amount = chipo.groupby('order_id')['item_revenue'].sum()
avg_revenue = orderwise_amount.mean()
print(f"Average Revenue Per Order : ${avg_revenue:.2f}")
```
